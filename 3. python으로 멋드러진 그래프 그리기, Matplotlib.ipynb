```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

## II. Matplotlib Case Study

```python
plt.plot([1, 2, 3, 4, 5])    # 실제 plotting을 하는 함수, y = x + 1
                             # 여기서 x=index, y=데이터
                             # plot 함수는 꺽은 선 그래프이다.
plt.show()    # plt를 확인하는 명령
```

### Figsize: Figure(도면)의 크기 조절

```python
plt.figure(figsize=(3, 3))   # plotting을 할 도면을 선언

plt.plot([0, 1, 2, 3, 4]) 
plt.show()
```

```python
plt.figure(figsize=(10, 10))   # plotting을 할 도면을 선언

plt.plot([0, 1, 2, 3, 4]) 
plt.show()
```

### 2차 함수 그래프 with plot()

```python
# 리스트를 이용해서 1차 함수 y=x를 그려보면:

plt.plot([0, 1, 2, 3, 4])
plt.show()
```

```python
# numpy.array를 이용해서 함수 그래프 그리기

x = np.array([1, 2, 3, 4, 5])      # 정의역
y = np.array([1, 4, 9, 16, 25])    # f(x)

plt.plot(x,y)
plt.show()
```

```python
# np.arange(a, b, c)

x = np.arange(-10, 10, 0.01)
x[:5]
```

```python
plt.plot(x, x**2)
plt.show()
```

```python
# x, y축에 설명 추가하기

x = np.arange(-10, 10, 0.01)

###
plt.xlabel("x value")
plt.ylabel("f(x) value")
###

plt.plot(x, x**2)
plt.show()
```

```python
# x, y축의 범위를 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

###
plt.axis([-5, 5, 0, 25])    # [x_min, x_max, y_min, y_max]
###

plt.plot(x, x**2)
plt.show()
```

```python
# x, y축에 눈금 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25])    # [x_min, x_max, y_min, y_max]

###
plt.xticks([i for i in range(-5, 6, 1)])    # x축 눈금 설정
plt.yticks([i for i in range(0, 27, 3)])    # y축 눈금 설정
###

plt.plot(x, x**2)
plt.show()
```

```python
# 그래프에 title 달기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25])    # [x_min, x_max, y_min, y_max]
plt.xticks([i for i in range(-5, 6, 1)])    # x축 눈금 설정
plt.yticks([i for i in range(0, 27, 3)])    # y축 눈금 설정

###
plt.title("y = x^2 graph")    # 그래프 title
###

plt.plot(x, x**2)
plt.show()

```

```python
# 그래프에 범례 사용

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25])    # [x_min, x_max, y_min, y_max]
plt.xticks([i for i in range(-5, 6, 1)])    # x축 눈금 설정
plt.yticks([i for i in range(0, 27, 3)])    # y축 눈금 설정
plt.title("y = x^2 graph")    # 그래프 title

###
plt.plot(x, x**2, label="trend")
plt.legend()
###

plt.show()

```

## III. Matplotlib Case Study

### 꺽은 선 그래프 (Plot)

- `.plot()`

```python
x = np.arange(20)    # 0~19
y = np.random.randint(0, 20, 20)    # 난수를 20번 생성

x, y
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(20)    # 0~19
y = np.random.randint(0, 20, 20)    # 난수를 20번 생성
plt.plot(x, y)

# Extra: y축을 20까지 보이게 하고 싶다면?
# y축을 "5"단위로 보이게 하고 싶다면?
# .axis(), .yticks()

plt.axis([0, 20, 0, 20])
plt.yticks([0,5,10,15,20])

plt.show()
```

### 산점도(Scatter Plot)

- `.scatter()`

```python
plt.scatter(x, y)
plt.show()
```

### 박스 그림 (Box Plot)

- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)
- `.boxplot()`

```python
plt.boxplot(y)
plt.show()
```

```python
# Extra: plot의 title을 "Box plot of y"

plt.boxplot(y)
plt.title("Box plot of y")
plt.show()
```

```python
# Extra: boxplot()함수의 인자를 컨테이너로 주어 두 변수를 비교하는 데 사용

plt.boxplot((x,y))
plt.title("Box plot of y")
plt.show()
```

### 막대 그래프 (Bar Plot)

- 범주형 데이터의 "값"과 그 값의 크기를 직사각형으로 나타낸 그림

```python
plt.bar(x, y)
plt.show()
```

```python
# Extra: xticks를 올바르게 처리해보자.

plt.bar(x, y)
plt.xticks(np.arange(0, 20, 1))
plt.show()
```

### 히스토그램 (histogram)

```python
# 도수분포를 직사각형의 막대 형태로 나타냈다.
# "계급"으로 나타낸 것이 특징: 0, 1, 2가 아니라 "범주형" 데이터로 구성 후 그림을 그림

plt.hist(y, bins=np.arange(0, 22, 2))    # bins = 범주의 간격
plt.show()
```

```python
# Extra: xticks를 올바르게 고쳐보자.

plt.hist(y, bins=np.arange(0, 22, 2))    # bins = 범주의 간격
plt.xticks(np.arange(0, 20, 2))
plt.show()
```

### 원형 그래프 (Pie Chart)

- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해서 **비율** 확인에 용이
- `.pie()`

```python
z = [100, 300, 200, 400]

plt.pie(z, labels=['one', 'two', 'three', 'four'])
plt.show()
```

### Seaborn Import 하기

```python
import seaborn as sns
```

### 커널밀도 그림 (Kernel Density Plot)
- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림
- sns.kdeplot()

```python
# in Histogram

x = np.arange(0, 22, 2)
y = np.random.randint(0,20,20)

plt.hist(y, bins=x)
plt.show()
```

```python
# kdeplot

sns.kdeplot(y)

plt.show()
```

```python
# kdeplot

sns.kdeplot(y, shade=True)

plt.show()
```

### 카운트 그림 (Count Plot)
- 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과
- `sns.countplot()`

```python
vote_df = pd.DataFrame({"name":["Andy", "Bob", "Cat"], "vote": [True, True, False]})

vote_df
```

```python
# in matplotlib barplot

vote_count = vote_df.groupby('vote').count()

vote_count
```

```python
plt.bar(x=[False, True], height=vote_count['name'])
plt.show()
```

```python
# sns의 countplot

sns.countplot(x=vote_df["vote"])
plt.show()
```

### 캣 그림 (Cat Plot)
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- `sns.catplot()`

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid.head(5)
```

```python
sns.catplot(x='WHO Region', y='Confirmed', data=covid)
plt.show()
```

```python
s= sns.catplot(x='WHO Region', y='Confirmed', data=covid)
s.fig.set_size_inches(10, 6)
plt.show()
```

```python
s= sns.catplot(x='WHO Region', y='Confirmed', data=covid, kind='violin')
s.fig.set_size_inches(10, 6)
plt.show()
```

### 스트립 그림 (Strip Plot)
- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- `sns.stripplot()`

```python
sns.stripplot(x='WHO Region', y='Recovered', data=covid)

plt.show()
```

```python
#cf) swarmplot

#점들이 겹치는 경우 점들을 양 옆으로 분산
# 따라서 데이터를 확인하는 것이 용이

sns.swarmplot(x='WHO Region', y='Recovered', data=covid)

plt.show()    # 오류가 발생한 이유는 주어진 점들의 값이 너무 커서(너무 겹쳐서) 다 그릴 수 없다는 의미이므로 현재는 무시해도 괜찮다.
```

### 히트맵 (Heatmap)
- 데이터의 행렬을 색상으로 표현해주는 그래프
- `sns.heatmap()`

```python
# 히트맵 예제

covid.corr()
```

```python
sns.heatmap(covid.corr())

plt.show()
```
